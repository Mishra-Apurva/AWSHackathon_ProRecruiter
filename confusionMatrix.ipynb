```python
import json
import matplotlib.pyplot as plt
import numpy as np

# Example JSON string
json_str = '{"confusion_matrix": [[5, 0, 0, 0, 0], [0, 2, 0, 1, 0], [0, 1, 6, 0, 0], [0, 0, 0, 2, 0], [0, 0, 0, 1, 3]], "labels": ["Cold emails", "Rejection", "acceptance ", "follow up", "interview acceptance "], "type": "multi_class", "all_labels": ["Cold emails", "Rejection", "acceptance ", "follow up", "interview acceptance ", "label"]}'

# Load the JSON string into a Python dictionary
confusion_dict = json.loads(json_str)

# Extract the confusion matrix and labels from the dictionary
confusion_matrix = np.array(confusion_dict['confusion_matrix'])
labels = confusion_dict['labels']

# Create a figure and axis for the plot
fig, ax = plt.subplots()

# Set the colormap
cmap = plt.cm.Blues

# Normalize the confusion matrix
cm_norm = confusion_matrix.astype('float') / confusion_matrix.sum(axis=1)[:, np.newaxis]

# Plot the matrix as an image
im = ax.imshow(cm_norm, interpolation='nearest', cmap=cmap)

# Add the colorbar
fig.colorbar(im)

# Set the tick marks and labels
ax.set_xticks(np.arange(len(labels)))
ax.set_yticks(np.arange(len(labels)))
ax.set_xticklabels(labels)
ax.set_yticklabels(labels)

# Rotate the tick labels and set their alignment
plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
         rotation_mode="anchor")

# Loop over data dimensions and create text annotations.
for i in range(len(labels)):
    for j in range(len(labels)):
        text = ax.text(j, i, confusion_matrix[i, j],
                       ha="center", va="center", color="black")

# Set the plot title and axis labels
ax.set_title("Confusion Matrix")
ax.set_xlabel("Predicted Label")
ax.set_ylabel("True Label")

# Display the plot
plt.show()
```
